```python
import torch
```

```python
x = torch.ones(5)
y = torch.zeros(3)
w = torch.rand(5, 3, requires_grad=True)
b = torch.rand(3, requires_grad=True)
```

```python
z = torch.matmul(x,w) + b
```

```python
loss = torch.nn.functional.binary_cross_entropy_with_logits(z, y)
```

```python
loss.backward()
```
